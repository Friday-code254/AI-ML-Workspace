# 7. Performance Metrics

A **performance metric** measures model quality.

For regression problems:

## RMSE (Root Mean Squared Error)

Most common metric.

Measures average prediction error.

---

## RMSE Formula

$$RMSE = \sqrt{\frac{1}{m}\sum_{i=1}^{m}(\hat{y}^{(i)} - y^{(i)})^2}$$

Where:

* (m) = number of examples
* (\hat{y}) = predicted value
* (y) = actual value

---

## Why RMSE Is Important

* Penalizes large errors heavily
* Lower RMSE = better model

Example:

* Error of 100 is much worse than error of 10

---

# 8. MAE (Mean Absolute Error)

Another regression metric.

## Formula

$$MAE = \frac{1}{m}\sum_{i=1}^{m}|\hat{y}^{(i)} - y^{(i)}|$$

---

## Difference Between RMSE and MAE

| Metric | Behavior                   |
| ------ | -------------------------- |
| RMSE   | Punishes large errors more |
| MAE    | Treats all errors equally  |

---

## When to Use MAE

Use MAE when:

* Dataset has many outliers
* You don’t want huge punishment for large errors
