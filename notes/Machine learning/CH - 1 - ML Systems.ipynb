# Types of Machine Learning Systems

Machine Learning systems are mainly classified based on:

1. **Training Supervision**
2. **Learning Style**
3. **Generalization Method**

---

# 1. Training Supervision Types

This classification depends on how much guidance or labeled data the model receives during training.

The main categories are:

- Supervised Learning
- Unsupervised Learning
- Semi-Supervised Learning
- Self-Supervised Learning
- Reinforcement Learning

---

# A. Supervised Learning

In **Supervised Learning**, the training dataset contains:

- **Inputs (Features)**
- **Correct Outputs (Labels/Targets)**

The model learns a mapping:

```text
Input → Output
```

The goal is:
- Given new unseen input,
- Predict the correct output.

---

## Main Tasks in Supervised Learning

### 1. Classification

The model predicts **categories/classes**.

#### Examples:
- Spam vs Not Spam
- Cat vs Dog
- Positive vs Negative Review
- Fraud vs Legit Transaction

#### Output Type:
- Discrete labels/categories

---

### 2. Regression

The model predicts **numerical values**.
Regression is a supervised learning technique used to predict continuous numerical values by learning relationships between input variables (features) and an output variable (target). It helps understand how changes in one or more factors influence a measurable outcome and is widely used in forecasting, risk analysis, decision-making and trend estimation.



#### Examples:
- House price prediction
- Salary prediction
- Stock price forecasting
- Temperature prediction

#### Output Type:
- Continuous numerical values

---

## Important Terms

| Term | Meaning |
|---|---|
| Feature | Input variable |
| Label/Target | Correct output |
| Classification | Predict categories |
| Regression | Predict numerical values |

---

## Supervised Learning Workflow

```text
Labeled Data
     ↓
Train Model
     ↓
Learn Patterns
     ↓
Predict on New Data
```

---

## Common Supervised Learning Algorithms

### Classification Algorithms
- Logistic Regression
- Decision Trees
- Random Forest
- Support Vector Machines (SVM)
- Neural Networks

### Regression Algorithms
- Linear Regression
- Polynomial Regression

- The model continuously learns from incoming data.

- Incremental learning
- Fast adaptation

- Efficient for huge datasets

---

```text
Mini-Batches
```

Controls:
- How fast the model updates itself.

- Adapts rapidly

- Corrupt model

# Solutions

# Main Challenges of Machine Learning

Machine Learning systems learn patterns from data.  
However, building a successful ML system is difficult because many things can go wrong during:

- data collection,
- training,
- evaluation,
- deployment.

In general, ML failures usually come from two major sources:

1. **Bad Data**
2. **Bad Models**

A model may fail because:
- the dataset is too small,
- the data is biased,
- the data contains noise,
- the features are irrelevant,
- the model is too complex,
- or the model is too simple.

---

# 1. Insufficient Quantity of Training Data

Machine Learning algorithms usually require a **large amount of training data** to work properly.

Humans can learn concepts from only a few examples.  
For example, a child may learn what an apple is after seeing only a few apples.

Machines are very different.

Most ML systems need:
- Thousands of examples for simple tasks
- Millions of examples for complex tasks like:
  - image recognition
  - speech recognition
  - natural language processing

---

## Why More Data Helps

When the dataset becomes larger:
- the model sees more variations,
- learns better patterns,
- and generalizes more effectively.

With enough data:
- even relatively simple algorithms can perform surprisingly well.

---

## The Unreasonable Effectiveness of Data

A famous research paper by:
- Michele Banko
- Eric Brill

showed that many ML algorithms performed almost equally well once they were given enough data.

This idea later became widely known through the paper:

**"The Unreasonable Effectiveness of Data"**

The central idea:

> For many complex problems, having more data is often more important than having a more sophisticated algorithm.

---

## Example

Suppose you want to build:
- a cat vs dog classifier.

### Small Dataset
If you train using:
- only 50 images,

the model may memorize the training images instead of learning general features.

### Large Dataset
If you train using:
- 5 million images,

the model can learn:
- fur textures,
- ear shapes,
- lighting conditions,
- different backgrounds,
- different camera angles.

This dramatically improves performance.

---

## Transfer Learning

Gathering huge datasets is expensive.

A common solution is:

# Transfer Learning

A model trained on one task reuses knowledge for another task.

Example:
A neural network trained on millions of images already understands:
- edges,
- shapes,
- textures,
- object structures.

It can later be fine-tuned for:
- flower classification,
- medical diagnosis,
- face recognition.

This reduces the amount of required training data.

---

# 2. Non-Representative Training Data

A model learns only from the data it sees.

If training data does not properly represent real-world cases,
the model will perform poorly in production.

---

# Representative Data

Training data should resemble:
- future real-world inputs.

If not:
- the model learns incorrect patterns,
- predictions become unreliable.

---

## Example: GDP vs Happiness

Suppose you train a model using countries whose GDP per capita lies only between:

- \$23,500
and
- \$62,500

The model never sees:
- extremely poor countries,
- extremely rich countries.

As a result:
- predictions for those countries become inaccurate.

The learned relationship becomes misleading.

---

# Sampling Noise

If the dataset is too small,
random chance may create misleading patterns.

This is called:

#### Sampling Noise

The data becomes accidentally unrepresentative.

---

# Sampling Bias

Sometimes the method used to collect data itself is flawed.

This is called:

#### Sampling Bias

The dataset systematically favors certain groups over others.

---

# Famous Example: 1936 US Election Poll

The magazine:
- The Literary Digest

conducted a huge election poll.

It predicted:
- Alf Landon would win.

But:
- Franklin D. Roosevelt won by a huge margin.

---

## Why the Prediction Failed

The magazine collected addresses using:
- telephone directories,
- magazine subscriptions,
- club membership lists.

At that time:
- wealthy people were more likely to own telephones,
- wealthy people leaned Republican.

So the sample was heavily biased.

---

# Non-Response Bias

Another problem:
many people ignored the survey.

People who answer surveys may differ from those who ignore them.

This creates:

#### Non-Response Bias

---

# Example in ML

Suppose you build:
- a funk music classifier

using only:
- YouTube search results.

Problem:
YouTube rankings are already biased toward:
- popular artists,
- trending content,
- local preferences.

So your dataset may not represent all funk music.

---

# 3. Poor-Quality Data

If training data contains:
- errors,
- missing values,
- noise,
- outliers,

then the model struggles to learn useful patterns.

---

# Noisy Data

Noise refers to:
- random errors,
- corrupted measurements,
- incorrect labels.

Examples:
- blurry images,
- mislabeled emails,
- incorrect customer ages.

---

# Outliers

Outliers are abnormal extreme values.

Example:
- customer age = 900 years.

Outliers can distort learning.

---

# Missing Features

Some records may have incomplete information.

Example:
- customers not specifying age.

Possible solutions:
- remove the feature,
- remove incomplete rows,
- fill missing values,
- train multiple models.

---

# Data Cleaning

Real-world ML requires extensive:

# Data Cleaning

Tasks include:
- fixing incorrect labels,
- removing duplicates,
- handling missing values,
- removing corrupted samples,
- correcting inconsistent formatting.

Data scientists often spend more time cleaning data than building models.

---

# 4. Irrelevant Features

Machine Learning depends heavily on:
- useful input features.

If irrelevant features dominate the dataset,
the model learns useless patterns.

This is summarized by the famous phrase:

## Garbage In → Garbage Out

---

Feature engineering is the process of creating useful inputs for ML models.

It is one of the most important parts of ML.

---

# Types of Feature Engineering

## A. Feature Selection

Choosing the most useful features.

Example:
For house-price prediction:
- area,
- location,
- number of bedrooms

matter more than:
- house ID.

---

## B. Feature Extraction

Combining features into better representations.

Example:
- car age
- mileage

can combine into:
- wear-and-tear score.

---

## C. Creating New Features

Gathering additional useful information.

Example:
For weather prediction:
- humidity,
- wind speed,
- atmospheric pressure

may improve performance.

---

# 5. Overfitting

Overfitting happens when:
- a model learns training data too perfectly,
including:
- noise,
- random fluctuations,
- accidental patterns.

The model performs:
- extremely well on training data,
- poorly on unseen data.

---

# Example of Overfitting

Suppose a model learns:

> Countries containing the letter "w" are happier.

Training examples:
- Norway
- Sweden
- Switzerland
- New Zealand

This pattern is meaningless in reality.

The model learned:
- random noise,
not:
- real relationships.

---

# Why Overfitting Happens

Overfitting usually occurs when:
- the model is too complex,
- the dataset is too small,
- the training data is noisy.

---

# Signs of Overfitting

| Dataset | Performance |
|---|---|
| Training Set | Excellent |
| Test Set | Poor |

---

# Solutions to Overfitting

## A. Simplify the Model

Use:
- linear regression
instead of:
- extremely complex polynomial models.

---

## B. Gather More Data

More data helps identify:
- real patterns,
instead of:
- random noise.

---

## C. Reduce Noise

Clean the dataset.

---

## D. Regularization

Regularization constrains model complexity.

It discourages large parameter values.

Example:

Linear Regression Equation:

y = θ₀ + θ₁x

If θ₁ becomes too large,
the model may overfit.

Regularization penalizes overly large weights.

---

# Hyperparameters

A hyperparameter controls:
- the learning process itself.

Examples:
- learning rate,
- regularization strength,
- tree depth.

Unlike normal parameters:
- hyperparameters are set before training.

---

# 6. Underfitting

Underfitting is the opposite of overfitting.

It happens when:
- the model is too simple
to capture underlying patterns.

---

# Example

Using a straight line to model highly nonlinear data.

The model fails:
- even on training data.

---

# Signs of Underfitting

| Dataset | Performance |
|---|---|
| Training Set | Poor |
| Test Set | Poor |

---

# Solutions to Underfitting

## A. Use a More Powerful Model

Example:
- neural networks instead of linear regression.

---

## B. Add Better Features

Feature engineering can improve learning.

---

## C. Reduce Regularization

Too much regularization may oversimplify the model.

---

# Testing and Validation

A trained model must be evaluated on unseen data.

---

# Train-Test Split

Dataset is divided into:

| Dataset | Purpose |
|---|---|
| Training Set | Learn patterns |
| Test Set | Evaluate performance |

Common split:
- 80% training
- 20% testing

---

# Generalization Error

The error on unseen data is called:

# Generalization Error

or:

# Out-of-Sample Error

It estimates:
- how well the model performs in the real world.

---

# Detecting Overfitting

If:
- training error is low,
- test error is high,

then the model is overfitting.

---

# Validation Set

Using the test set repeatedly causes hidden overfitting.

Solution:
split data into:
- training set,
- validation set,
- test set.

---

# Holdout Validation

Workflow:

| Step | Action |
|---|---|
| 1 | Train on training set |
| 2 | Evaluate on validation set |
| 3 | Tune hyperparameters |
| 4 | Final evaluation on test set |

---

# Cross-Validation

Instead of using one validation split:

- divide data multiple times,
- train repeatedly,
- average the results.

This gives:
- more reliable evaluation.

Tradeoff:
- higher computational cost.

---

# Data Mismatch

Sometimes training data differs from production data.

---

# Example

Suppose you build:
- a flower recognition app.

Training images:
- high-quality web images.

Production images:
- blurry mobile camera photos.

The model may fail because:
- training distribution ≠ real-world distribution.

---

# Train-Dev Set

Andrew Ng proposed:

# Train-Dev Set

Purpose:
- distinguish between:
  - overfitting,
  - and data mismatch.

---

# Workflow

| Dataset | Purpose |
|---|---|
| Training Set | Learn |
| Train-Dev Set | Detect overfitting |
| Validation Set | Detect data mismatch |
| Test Set | Final evaluation |

---

# No Free Lunch Theorem

There is:

> No universally best machine learning algorithm.

Different datasets require different models.

---

# Core Idea

Every model makes assumptions about data.

Example:
Linear Regression assumes:
- relationships are approximately linear.

But many real-world problems are nonlinear.

---

# Implication

No single model works best everywhere.

Sometimes:
- linear regression works best.

Other times:
- neural networks,
- decision trees,
- SVMs,
- ensemble methods
perform better.

---

# Practical ML Strategy

In practice:
- make reasonable assumptions,
- test multiple models,
- compare performance,
- tune carefully.
