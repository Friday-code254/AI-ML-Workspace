### Pandas
It is a powerful data maniplulation libray in python, widly used for for analysis and data cleaning. it has two primary data structures: series and DataFrams. Series - one-dimensional array like objects. DataFrames - two-dimensional, size-mutable, and potentially heterogenous tabular data structure with labled axes(rows and columns)

```python
import pandas as pd
```

```python
### series
### it is a one dimensional array with index and values
### it can be created from list, dict, numpy array
data = [1,2,3,4,5,6]
series = pd.Series(data)
print(series)

dict = {'a':1, 'b':2, 'c':3}
series = pd.Series(dict)
print(series)

data = [1,2,3,4,5,6]
i = ['a','b','c','d','e','f']   
series = pd.Series(data, index = i )
print(series)
```

```python
group_mean = dfc.groupby('PRODUCTLINE')['SALES'].mean()
group_mean### it will group the data by PRODUCTLINE and calculate the mean of sales for each PRODUCTLINE

group_sum = dfc.groupby(['PRODUCTLINE', 'COUNTRY'])['SALES'].sum()
group_sum### it will group the data by PRODUCTLINE and COUNTRY and calculate the sum of sales for each group

group_agg = dfc.groupby('PRODUCTLINE')['SALES'].agg(['mean', 'sum', 'count']) ### it will group the data by PRODUCTLINE and calculate the mean, sum and count of sales for each PRODUCTLINE
group_agg
```

```python
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
print(pd.merge(df1,df2 ,on='Key' ,how= 'inner'), "\n")### it will merge the two dataframes based on the common column 'Key' and return only the rows that have matching keys in both dataframes
print(pd.merge(df1,df2 ,on='Key' ,how= 'outer'), "\n")### it will  merge the two dataframes based on the common column 'Key' and return all the rows from both dataframes, filling in NaN for missing values
print(pd.merge(df1,df2 ,on='Key' ,how= 'left'), "\n")### it will merge the two dataframes based on the common column 'Key' and return all the rows from the left dataframe, filling in NaN for missing values in the right dataframe
print(pd.merge(df1,df2 ,on='Key' ,how= 'right'), "\n")### it will merge the two dataframes based on the common column 'Key' and return all the rows from the right dataframe, filling in NaN for missing values in the left dataframe
print(pd.merge(df1, df2, how='cross'), "\n")### it will merge the two dataframes based on the common column 'Key' and return all the possible combinations of rows from both dataframes
```
