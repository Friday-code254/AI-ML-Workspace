### SQL And SQLite

SQL. (Structured Query Language) is a standard language for managing and manipulating relational databases.
SQLite is a self-contained, serverless, and zero-configuration database engine that is widely used for embedded database systems. 

```python
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
cursor  = connection.cursor() 
```

```python
cursor.execute('''
Create Table If Not Exists emplyees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department Text
    )   
''')
connection.commit()
```

```python
cursor.execute('''
select * from emplyees''') ###
```

```python
cursor.execute('''
insert into emplyees(name,age,department) values('Alice',30,'HR')
''')

cursor.execute('''
insert into emplyees(name,age,department) values('Bob',25,'IT')''')

cursor.execute('''
insert into emplyees(name,age,department) values('Charlie',28,'Finance')''')

## or

data = [
    ('Alice', 30, 'HR'),
    ('Bob', 25, 'IT'),
    ('Charlie', 28, 'Finance')
]

cursor.executemany('''
insert into emplyees(name,age,department) values(?,?,?)
''', data)

connection.commit()
```

```python
cursor.execute(''' 
select * from emplyees''')

rows = cursor.fetchall() ### it fecthes all the rows from the result of the query and returns them as a list of tuples.
for row in rows:
    print(row)
```

```python
cursor.execute('''
update emplyees 
set age = 31 
where name = 'Alice' 
''')
connection.commit()
```

```python
cursor.execute('''
delete from emplyees
where name = 'Bob'
 ''')
connection.commit()
```

```python
connection.close() 
```
