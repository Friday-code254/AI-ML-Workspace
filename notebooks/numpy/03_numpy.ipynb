### NumPy
is a core Python library for numerical computing, built for handling large arrays and matrices efficiently.

```python
import numpy as np

a = [9,3,3,5]
print(np.array(a))  # Converting list into numpy array
zero = np.zeros((3,4))  # Creating a 3x4 array filled with zeros
print(zero)
one = np.ones((2,3))  # Creating a 2x3 array filled with ones
print(one)
ar = np.arange(1,10,2)  # Creating an array with values from 1 to 9 with a step of 2
print(ar)   
random_array = np.random.rand(5,3)  # Generating an array of 5 random numbers between 0 and 1
print(random_array)
```
